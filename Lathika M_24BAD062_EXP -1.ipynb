```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
print("Name     : Lathika M")
print("Roll No  : 24BAD062")
print("Scenario : SCENARIO 1 â€“ E-commerce Sales Data Analysis\n")
df = pd.read_csv("/kaggle/input/ecommerce-data/data.csv", encoding="ISO-8859-1")
print("----- HEAD -----")
print(df.head(), "\n")
print("----- TAIL -----")
print(df.tail(), "\n")
print("----- INFO -----")
df.info()
print("\n")
print("----- DESCRIBE -----")
print(df.describe(), "\n")
print("----- MISSING VALUES -----")
print(df.isnull().sum(), "\n")
df['Sales'] = df['Quantity'] * df['UnitPrice']
df = df[(df['Quantity'] > 0) & (df['UnitPrice'] > 0)]
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
top5_products = (
    df.groupby('Description')['Sales']
    .sum()
    .sort_values(ascending=False)
    .head(5)
)
plt.figure()
top5_products.plot(kind='bar')
plt.title("Top 5 Products by Sales")
plt.xlabel("Product")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.show()
df_2011 = df[df['InvoiceDate'].dt.year == 2011]
monthly_sales_2011 = (
    df_2011.groupby(df_2011['InvoiceDate'].dt.month)['Sales']
    .sum()
)
plt.figure()
monthly_sales_2011.plot()
plt.title("Monthly Sales Trend in 2011")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.xticks(range(1, 13))
plt.show()
```

```python
print("Name     : Lathika M")
print("Roll No  : 24BAD062")
print("Scenario : SCENARIO 3 â€“ Housing Dataset Analysis\n")
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("/kaggle/input/housing-price-prediction/Housing.csv")
print("----- DATASET COLUMNS -----")
print(df.columns, "\n"
print("----- FIRST 5 ROWS -----")
print(df.head(), "\n")
print("----- DATASET INFO -----")
df.info()
print("\n")
print("----- STATISTICAL SUMMARY -----")
print(df.describe(), "\n")
print("----- MISSING VALUES SUMMARY -----")
print(df.isnull().sum(), "\n")
plt.figure()
plt.scatter(df['area'], df['price'])
plt.title("House Price vs Area")
plt.xlabel("Area")
plt.ylabel("Price")
plt.show()
plt.figure()
plt.scatter(df['bedrooms'], df['price'])
plt.title("House Price vs Bedrooms")
plt.xlabel("Bedrooms")
plt.ylabel("Price")
plt.show()

plt.figure(figsize=(10, 8))
correlation_matrix = df.corr(numeric_only=True)
sns.heatmap(correlation_matrix, annot=True)
plt.title("Correlation Heatmap of Housing Features")
plt.show()

print("----- INTERPRETATION FOR PRICE PREDICTION -----")
print("1. House price has a strong positive relationship with area.")
print("2. Bedrooms and bathrooms show moderate correlation with price.")
print("3. Parking and stories also influence housing price.")
print("4. These features are important for predicting house prices.")
```

```python
print("Name       : Lathika M")
print("Roll No    : 24BAD062")
print("Scenario   : SCENARIO 4 â€“ Customer Personality Analysis")
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("/kaggle/input/customer-personality-analysis/marketing_campaign.csv", sep='\t')

print("----- COLUMNS -----")
print(df.columns, "\n")

print("----- FIRST 5 ROWS -----")
print(df.head(), "\n")

print("----- DATASET INFO -----")
df.info()
print("\n")

print("----- STATISTICAL SUMMARY -----")
print(df.describe(), "\n")

print("----- MISSING VALUES -----")
print(df.isnull().sum(), "\n")

df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'], errors='coerce')
df['Total_Spending'] = (
    df['MntWines'] + df['MntFruits'] + df['MntMeatProducts'] +
    df['MntFishProducts'] + df['MntSweetProducts'] + df['MntGoldProds']
)
plt.figure()
plt.hist(df['Year_Birth'], bins=20)
plt.title("Distribution of Customer Birth Years")
plt.xlabel("Year of Birth")
plt.ylabel("Frequency")
plt.show()
plt.figure()
plt.boxplot(df['Total_Spending'], vert=False)
plt.title("Boxplot of Total Customer Spending")
plt.xlabel("Total Spending")
plt.show()

education_spending = df.groupby('Education')['Total_Spending'].mean()

plt.figure()
education_spending.plot(kind='bar')
plt.title("Average Spending by Education Level")
plt.xlabel("Education")
plt.ylabel("Average Spending")
plt.show()

plt.figure(figsize=(10, 8))
corr = df[['Income', 'Total_Spending', 'Recency', 'NumWebPurchases',
           'NumCatalogPurchases', 'NumStorePurchases']].corr()
sns.heatmap(corr, annot=True)
plt.title("Correlation Heatmap of Customer Metrics")
plt.show()

print("----- INTERPRETATION -----")
print("1. Higher income customers tend to spend more.")
print("2. Education level influences customer spending behavior.")
print("3. Web and store purchases show strong relation with total spending.")
print("4. These patterns help identify the companyâ€™s ideal customers.")
```
